```python
import pandas as pd
```

```python
df_1 = pd.read_csv("amz_com-ecommerce_sample.csv",encoding= 'unicode_escape',usecols = ["product_name","retail_price","discounted_price"])
```

```python
df_1.head()
```

```python
df_1.columns = df_1.columns.str.replace('product_name', 'product name in flipkart')
df_1.columns = df_1.columns.str.replace('retail_price', 'retail price in flipkart')
df_1.columns = df_1.columns.str.replace('discounted_price', 'discounted price in flipkart')
```

```python
df_1.head()
```

```python
df_2 = pd.read_csv("flipkart_com-ecommerce_sample.csv",usecols = ["product_name","retail_price","discounted_price"])
```

```python
df_2.head()
```

```python
df_2.rename(columns = {'product_name':'product name in amazon', 'retail_price':'retail price in amazon',
                              'discounted_price':'discounted price in amazon'}, inplace = True)
```

```python
df_2.head()
```

```python
## First Way

df_3 = pd.merge(df_1,df_2,left_on='product name in flipkart',right_on='product name in amazon')
```

```python
df_3.head()
```

```python
df_3.shape
```

```python
## Second Way

df_4 = pd.merge(df_1,df_2,how='inner',left_on=['product name in flipkart','retail price in flipkart'],right_on=['product name in amazon','retail price in amazon'])
```

```python
df_4.head()
```

```python
df_4.shape
```

```python
## Third Way

df_5 = pd.merge(df_1,df_2,how='inner',left_on=['product name in flipkart','discounted price in flipkart'],right_on=['product name in amazon','discounted price in amazon'])
```
